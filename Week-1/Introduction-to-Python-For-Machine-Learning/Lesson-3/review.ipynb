Which year had the least correlation with ‘Element Code’? 

```python
years = data[["Y2014", "Y2015", "Y2016", "Y2017"]]
element = data["Element Code"]
years.corrwith(element).sort_values()
# x = pd.DataFrame(years.corrwith(element), columns=["corr"])
# x["corr"].sort_values()
```

Select columns ‘Y2017’ and ‘Area’, Perform a groupby operation on ‘Area’.  Which of these Areas had the 7th lowest sum in 2017?

```python
area_2017 = data[["Y2017", "Area"]]
gp = area_2017.groupby("Area")
gp.sum().sort_values(by="Y2017").head(7)
```
